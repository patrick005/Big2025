# 교제 84P csv 파일 변환용 코드

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/home/aa/Big2025/orange3/basket_analysis.csv')
df.head(3)
```

```python
df = df.drop('Unnamed: 0', axis=1) # Unnamed: 0 컬럼 제거
df1 = df.replace(True, 1).replace(False, "") # True를 1로, False를 결측값으로 변환
df1.head(3)
```

```python
# 바꾼 csv파일을 저장
df1.to_csv('/home/aa/Big2025/orange3/basket_analysis_number.csv', index=False)
```
